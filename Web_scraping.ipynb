```python
## import libraries
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
# read html page downloaded using selenium
with open('allel_links.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
# Run documnent through BeautifulSoup Object.
soup = BeautifulSoup(html,'lxml')
```

```python
## Extract all the tables from html file.
container =soup.find_all('table',{'class':'tblNormal'})
```

```python
# Check 2 table from html page.
container[1]
```

```python
## create empty list
link_list = []
# Retrive all hyperlink from table 2
for i in container[1]:
    #print(i.find_all("a"))
    table_con = i.find_all("a")
type(table_con)
```

```python
type(table_con)
```

```python
# from tble con all list of link will be created.
for i in table_con: 
    link_list.append(str(i))
    #print(i)
    
```

```python
# crete empty list to create proper link  
final_list=[]
for i in link_list:
    final_list.append('http://www.allelefrequencies.net/'+ i.split('">')[0][9:])
```

# retirive 2nd link

```python
import requests
from lxml import html
```

```python
# List to store the 2nd links from each page
link21_list = []
```

```python
# Loop through each link in link_list
for link in final_list:
    # Send a request to the page and get the HTML content
    response = requests.get(link)
    content = response.content

    # Parse the HTML content using lxml
    parsed_content = html.fromstring(content)

    # Get all links on the page
    all_links = parsed_content.xpath('//a/@href')
    #print(len(all_links))
    # Check if the page has at least 21 links
    if len(all_links) >= 21:
        if all_links[18].startswith('hla'):
        # Get the 21st link and append it to link21_list
            link21_list.append('http://www.allelefrequencies.net/'+str(all_links[18]))
        else:
            # If there are fewer than 21 links, append an empty string to link21_list
            link21_list.append('http://www.allelefrequencies.net/'+str(all_links[19]))
    else:
        pass

# Print the list of 21st links
#print(link21_list)
```

# create table 

```python
# List to store DataFrames for each page
df_list = []
```

```python
# Loop through each link in link_list
for link in link21_list:
    # Send a request to the page and get the HTML content
    response = requests.get(link)
    content = response.content

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(content, 'html.parser')

    # Find the table on the page
    tables = soup.find('table',{'class':'tblNormal'})
    #print(tables)
    
    # Convert the table to a pandas DataFrame
    if tables:
        df_list.append(pd.read_html(str(tables))[0])
    # If the table does not exist, print a message and pass
    else:
        print("No table found on the page: ", link)
        continue 
                 

# Concatenate all the DataFrames into one DataFrame
final_df = pd.concat(df_list, ignore_index=True)

# Print the final DataFrame
print(final_df)
```
